```python

texts = [
    "Rahul Gandhi addressed the crowd in Delhi during the Bharat Jodo Yatra.",
    "Google announced the Pixel 8 launch event in San Francisco on Monday.",
    "Rohit Sharma hit a double century against Sri Lanka in Chennai."
]

true_entities = [
    [("Rahul Gandhi", "PERSON"), ("Delhi", "GPE"), ("Bharat Jodo Yatra", "ORG")],
    [("Google", "ORG"), ("Pixel 8", "PRODUCT"), ("San Francisco", "GPE"), ("Monday", "DATE")],
    [("Rohit Sharma", "PERSON"), ("Sri Lanka", "GPE"), ("Chennai", "GPE")]
]
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
predicted_entities = []

for text in texts:
    doc = nlp(text)
    predicted_entities.append([(ent.text, ent.label_) for ent in doc.ents])

# Display predictions
predicted_entities
```

```python
def get_token_labels(text, entities):
    text_tokens = text.split()
    labels = ["O"] * len(text_tokens)

    for ent_text, ent_label in entities:
        ent_tokens = ent_text.split()

        for i in range(len(text_tokens)):
            if text_tokens[i:i + len(ent_tokens)] == ent_tokens:
                for j in range(len(ent_tokens)):
                    labels[i + j] = ent_label

    return labels
```

```python
y_true = []
y_pred = []

for i, text in enumerate(texts):
    y_true.extend(get_token_labels(text, true_entities[i]))
    y_pred.extend(get_token_labels(text, predicted_entities[i]))

# Check sample labels
print(y_true[:20])
print(y_pred[:20])
```

```python
from sklearn.metrics import classification_report, accuracy_score

print("Accuracy:", accuracy_score(y_true, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_true, y_pred))
```
